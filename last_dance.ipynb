# Özellik Seçimi

Neden Özellik Seçimi Yapıyoruz?

Modelin daha hızlı ve verimli çalışmasını sağlamak

Gereksiz, etkisiz veya fazla korelasyonlu değişkenleri elemek

Aşırı öğrenme (overfitting) riskini azaltmak

Daha az ama daha anlamlı veriyle yüksek doğruluk elde etmek

## SelectKBest

Her özelliğin hedef değişkenle olan ilişkisini istatistiksel olarak test eder.

En yüksek F-score değerine sahip K adet özelliği seçer.

Özellikle hedef değişken sürekli veya sıralıysa etkilidir.

```python
from sklearn.feature_selection import SelectKBest, f_classif

def select_k_best_with_scores(X_train_scaled, y_train, k):
    selector = SelectKBest(score_func=f_classif, k=k)
    selector.fit(X_train_scaled, y_train)

    scores = selector.scores_

    feature_scores = pd.DataFrame({
        'Feature': X_train_scaled.columns,
        'Score': scores,
    }).sort_values(by='Score', ascending=False)

    return feature_scores
```

```python
kbest_result = select_k_best_with_scores(X_train_scaled, y_train, k=5)
print(kbest_result)
```

Select Kbest özellik seçimindeki score'ları göz önünde bulundurarak yeni veri seti oluşturdum.


```python
# SelectKBest ile seçilen sütunları aldım
selected_features_kbest = ['AptitudeTestScore', 'HSC_Marks',
                          'ExtracurricularActivities', 'Projects',
                          'SSC_Marks']

# Bu özellik setini ayrı bir X tanımıyla sakladım
X_kbest = X_train_scaled[selected_features_kbest]
```

## RFE Yöntemi

Belirli bir tahmin modeli (örneğin, Logistic Regression) üzerinden çalışır.

Önemsiz bulduğu özellikleri yinelemeli olarak kaldırır.

En iyi performansı veren alt özellik kümesini arar.

Aşamalı olarak özellik çıkarımı yaptığı için güçlüdür.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

def rfe_feature_selection_with_table(X, y, n):
    # 1. Model tanımla
    model = LogisticRegression(max_iter=1000)

    # 2. RFE ile seçici oluştur
    selector = RFE(estimator=model, n_features_to_select=n)
    selector.fit(X, y)

    # 3. 'Selected' sütunu olmadan sadece sıralama içeren tablo
    ranking_df = pd.DataFrame({
        'Feature': X.columns,
        'Ranking': selector.ranking_
    }).sort_values(by='Ranking')



    return ranking_df
```

```python
rfe_table = rfe_feature_selection_with_table(X_train_scaled, y_train, n=1)

# Tabloyu yazdır
print(rfe_table)
```

```python
X_rfe = X_train_scaled[['AptitudeTestScore', 'HSC_Marks', 'SoftSkillsRating', 'CGPA', 'SSC_Marks']]
```

##  Mutual Information

Özelliklerin hedef değişkenle olan bilgi kazancı ilişkisini ölçer.

Özellikler ve hedef değişken arasındaki bağımlılığı yakalar.

Doğrusal olmayan ilişkileri bile tespit edebilir.

```python
from sklearn.feature_selection import mutual_info_classif

def mi_feature_selection_with_table(X, y):
    # MI skorlarını hesapla
    mi_scores = mutual_info_classif(X, y, random_state=42)
    
    # DataFrame'e çevir ve sırala
    mi_df = pd.DataFrame({
        'Feature': X.columns,
        'MI Score': mi_scores
    }).sort_values(by='MI Score', ascending=False)
    
    return mi_df
```


| MI Skoru | Anlamı                      |
| -------- | --------------------------- |
| 0.0      | Bağımsız (hiç bilgi vermez) |
| 0.01–0.1 | Zayıf ilişki                |
| 0.1–0.3  | Orta düzey ilişki           |
| >0.3     | Güçlü ilişki                |


```python
mi_table = mi_feature_selection_with_table(X_train_scaled, y_train)


# Sonuçları yazdır
print(mi_table)
```

# Model Eğitimi

Neden Birden Fazla Model Kullandık?

Farklı algoritmalar farklı veri yapılarında daha iyi sonuç verebilir.

Modelleri karşılaştırarak en yüksek başarıyı göstereni tespit etmek istedik.

Gerçek dünyada veri çeşitliliği fazla olduğu için çoklu test önemli.

| Sıra | Model Adı                        | Tür / Yaklaşım                     | Açıklama                                                                                                                                                 |
| ---- | -------------------------------- | ---------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------- |
| 1    | **Logistic Regression**          | Lineer / Parametrik                | Basit, hızlı ve yorumlaması kolay bir modeldir. Doğrusal sınırlar üzerinden çalışır. Özellikle temel sınıflandırma problemleri için uygundur.            |
| 2    | **Decision Tree**                | Karar Ağacı / Kural Bazlı          | Veriyi if-else şeklinde dallara ayırarak sınıflandırır. Kolay yorumlanabilir. Ancak aşırı öğrenmeye yatkındır.                                           |
| 3    | **Random Forest**                | Ensemble / Bagging                 | Birden fazla karar ağacının ortalamasını alarak tahmin yapar. Genellikle yüksek doğruluk sağlar. Overfitting riskini azaltır.                            |
| 4    | **Support Vector Machine (SVM)** | Margin-Maximizing / Kernel Tabanlı | Karmaşık sınıflandırma problemleri için güçlüdür. Sınıflar arası en iyi ayrım çizgisini (hiperdüzlem) bulmaya çalışır. Küçük veri kümelerinde etkilidir. |


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Özellik setlerini tanımla
X_sets = {'SelectKBest': X_kbest, 'RFE': X_rfe, 'Mutual Information': X_mi}

# Model yapılandırmaları
classifiers = {
    'Logistic Regression': LogisticRegression(
        C=1.0,                        
        max_iter=1000,               
        penalty='l2',
        class_weight='balanced',     
        solver='lbfgs',              
        random_state=42
    ),

    'Decision Tree': DecisionTreeClassifier(
        max_depth=None,              
        min_samples_split=2,        
        min_samples_leaf=1,         
        class_weight='balanced',
        random_state=42
    ),

    'Random Forest': RandomForestClassifier(
        n_estimators=100,            
        max_depth=None,              
        min_samples_split=2,
        min_samples_leaf=1,
        class_weight='balanced',
        n_jobs=-1,                  
        random_state=42
    ),

    'SVM': SVC(
        C=1.0,                       
        kernel='rbf',
        gamma='scale',               
        probability=True,            
        class_weight='balanced',
        random_state=42
    )
}

# Değerlendirme fonksiyonu
def evaluate_models(X_sets, y, classifiers, test_size=0.2, random_state=42):
    results = []

    for set_name, X in X_sets.items():
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

        for model_name, model in classifiers.items():
            model.fit(X_train, y_train)

            # Tahminler
            y_train_pred = model.predict(X_train)
            y_test_pred = model.predict(X_test)

            # 🔎 Metrikleri hesapla
            for split, y_true, y_pred in [('Train', y_train, y_train_pred), ('Test', y_test, y_test_pred)]:
                results.append({
                    'Feature Set': set_name,
                    'Model': model_name,
                    'Data Split': split,
                    'Accuracy': accuracy_score(y_true, y_pred),
                    'Precision': precision_score(y_true, y_pred, average='weighted', zero_division=0),
                    'Recall': recall_score(y_true, y_pred, average='weighted'),
                    'F1-Score': f1_score(y_true, y_pred, average='weighted')
                })

    return pd.DataFrame(results)

# Fonksiyonu çalıştır
results_df = evaluate_models(X_sets, y_train, classifiers)
print("\n--- Model Performans Bilgileri (Train & Test) ---")
print(results_df)
```

```python
# En iyi modeli bul
best_accuracy = 0
best_model = None
best_feature_set = None
best_split = None

# Test seti sonuçlarına göre en iyi modeli bul
test_results = results_df[results_df['Data Split'] == 'Test']

for _, row in test_results.iterrows():
    if row['Accuracy'] > best_accuracy:
        best_accuracy = row['Accuracy']
        best_model = row['Model']
        best_feature_set = row['Feature Set']
        best_split = row['Data Split']

print("\n=== En İyi Model Performans Bilgileri ===")
print(f"Model: {best_model}")
print(f"Özellik Seti: {best_feature_set}")
print(f"Veri Bölümü: {best_split}")
print(f"En Yüksek Doğruluk: {best_accuracy:.4f}")

# En iyi modelin detaylı metriklerini göster
best_model_metrics = test_results[
    (test_results['Model'] == best_model) & 
    (test_results['Feature Set'] == best_feature_set)
].iloc[0]

print("\nDetaylı Metrikler:")
print(f"Precision: {best_model_metrics['Precision']:.4f}")
print(f"Recall: {best_model_metrics['Recall']:.4f}")
print(f"F1-Score: {best_model_metrics['F1-Score']:.4f}")
```

```python
# Performans karşılaştırma görsellerini oluştur
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

# Test seti sonuçlarını filtrele
test_results = results_df[results_df['Data Split'] == 'Test']

for metric in metrics:
    plt.figure(figsize=(14, 8))
    sns.barplot(x='Model', y=metric, hue='Feature Set', data=test_results)
    plt.title(f'Model Performans Karşılaştırması - {metric}')
    plt.ylim(0.5, 1.0)
    plt.xticks(rotation=45)
    plt.legend(title='Özellik Seti', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# ROC Analizi için gerekli fonksiyonlar
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
import numpy as np

def plot_roc_curves(X_sets, y, classifiers, test_size=0.2, random_state=42):
    plt.figure(figsize=(15, 10))
    
    # Sınıf sayısını belirle
    classes = np.unique(y)
    n_classes = len(classes)
    
    for set_name, X in X_sets.items():
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
        
        # Çok sınıflı etiketleri binary formata dönüştür
        y_test_bin = label_binarize(y_test, classes=classes)
        
        for model_name, model in classifiers.items():
            model.fit(X_train, y_train)
            
            # Her sınıf için ROC eğrisi
            if hasattr(model, "predict_proba"):
                y_pred_proba = model.predict_proba(X_test)
                
                # Binary sınıflandırma durumu için özel işlem
                if n_classes == 2:
                    # Sadece pozitif sınıf olasılıklarını al
                    y_pred_proba = y_pred_proba[:, 1].reshape(-1, 1)
                    y_test_bin = y_test_bin.reshape(-1, 1)
                
                # Her sınıf için ROC eğrisi çiz
                for i in range(y_test_bin.shape[1]):
                    fpr, tpr, _ = roc_curve(y_test_bin[:, i], 
                                          y_pred_proba[:, i] if n_classes > 2 else y_pred_proba[:, 0])
                    roc_auc = auc(fpr, tpr)
                    plt.plot(fpr, tpr, 
                            label=f'{model_name} ({set_name}) - Class {classes[i]} - AUC: {roc_auc:.3f}')
            else:  # SVM için decision_function kullan
                y_pred_proba = model.decision_function(X_test)
                
                # Binary sınıflandırma durumu için özel işlem
                if y_pred_proba.ndim == 1:
                    y_pred_proba = y_pred_proba.reshape(-1, 1)
                    
                # Her sınıf için ROC eğrisi çiz
                for i in range(y_test_bin.shape[1]):
                    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
                    roc_auc = auc(fpr, tpr)
                    plt.plot(fpr, tpr, 
                            label=f'{model_name} ({set_name}) - Class {classes[i]} - AUC: {roc_auc:.3f}')
    
    plt.plot([0, 1], [0, 1], 'k--')  # Referans çizgisi
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Eğrileri Karşılaştırması')
    plt.legend(loc="lower right", bbox_to_anchor=(1.15, 0))
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ROC analizini gerçekleştir
plot_roc_curves(X_sets, y_train, classifiers)
```

```python
import pickle
import os
from sklearn.model_selection import train_test_split

# En iyi model ve özellik seti tanımları
best_model_name = 'SVM'
best_feature_selection = 'Mutual Information'

# Model ve veri setini al
best_clf = classifiers[best_model_name]
X_best = X_sets[best_feature_selection]

# Eğitim/test bölünmesi (tekrar yapılır çünkü evaluate_models içinde lokal yapılıyor)
X_train_best, X_test_best, y_train_best, y_test_best = train_test_split(
    X_best, y_train, test_size=0.2, random_state=42
)

# Modeli eğit
best_clf.fit(X_train_best, y_train_best)

# Kaydedilecek dosya adı (dinamik olarak oluşturuluyor)
model_dir = 'model'
model_filename = f"model_{best_model_name.replace(' ', '')}_{best_feature_selection.replace(' ', '')}.pkl"
model_path = os.path.join(model_dir, model_filename)

# Klasör oluşturulmamışsa oluştur
os.makedirs(model_dir, exist_ok=True)

# Modeli ve gerekli bilgileri kaydet
with open(model_path, 'wb') as f:
    pickle.dump({
        'model': best_clf,
        'feature_selection_method': best_feature_selection,
        'model_name': best_model_name,
        'feature_names': X_best.columns.tolist(),
    
    }, f)

print(f"\n✅ En iyi model kaydedildi: {model_path}")
print(f"Özellik isimleri: {X_best.columns.tolist()}")
```
