## SelectKBest

Her Ã¶zelliÄin hedef deÄiÅkenle olan iliÅkisini istatistiksel olarak test eder.

En yÃ¼ksek F-score deÄerine sahip K adet Ã¶zelliÄi seÃ§er.

Ãzellikle hedef deÄiÅken sÃ¼rekli veya sÄ±ralÄ±ysa etkilidir.

```python
from sklearn.feature_selection import SelectKBest, f_classif

def select_k_best_with_scores(X_train_scaled, y_train, k):
    selector = SelectKBest(score_func=f_classif, k=k)
    selector.fit(X_train_scaled, y_train)

    scores = selector.scores_

    feature_scores = pd.DataFrame({
        'Feature': X_train_scaled.columns,
        'Score': scores,
    }).sort_values(by='Score', ascending=False)

    return feature_scores
```

```python
kbest_result = select_k_best_with_scores(X_train_scaled, y_train, k=5)
print(kbest_result)
```

Select Kbest Ã¶zellik seÃ§imindeki score'larÄ± gÃ¶z Ã¶nÃ¼nde bulundurarak yeni veri seti oluÅturdum.


```python
# SelectKBest ile seÃ§ilen sÃ¼tunlarÄ± aldÄ±m
selected_features_kbest = ['AptitudeTestScore', 'HSC_Marks',
                          'ExtracurricularActivities', 'Projects',
                          'SSC_Marks']

# Bu Ã¶zellik setini ayrÄ± bir X tanÄ±mÄ±yla sakladÄ±m
X_kbest = X_train_scaled[selected_features_kbest]
```

## RFE YÃ¶ntemi

Belirli bir tahmin modeli (Ã¶rneÄin, Logistic Regression) Ã¼zerinden Ã§alÄ±ÅÄ±r.

Ãnemsiz bulduÄu Ã¶zellikleri yinelemeli olarak kaldÄ±rÄ±r.

En iyi performansÄ± veren alt Ã¶zellik kÃ¼mesini arar.

AÅamalÄ± olarak Ã¶zellik Ã§Ä±karÄ±mÄ± yaptÄ±ÄÄ± iÃ§in gÃ¼Ã§lÃ¼dÃ¼r.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

def rfe_feature_selection_with_table(X, y, n):
    # 1. Model tanÄ±mla
    model = LogisticRegression(max_iter=1000)

    # 2. RFE ile seÃ§ici oluÅtur
    selector = RFE(estimator=model, n_features_to_select=n)
    selector.fit(X, y)

    # 3. 'Selected' sÃ¼tunu olmadan sadece sÄ±ralama iÃ§eren tablo
    ranking_df = pd.DataFrame({
        'Feature': X.columns,
        'Ranking': selector.ranking_
    }).sort_values(by='Ranking')



    return ranking_df
```

```python
rfe_table = rfe_feature_selection_with_table(X_train_scaled, y_train, n=1)

# Tabloyu yazdÄ±r
print(rfe_table)
```

##  Mutual Information

Ãzelliklerin hedef deÄiÅkenle olan bilgi kazancÄ± iliÅkisini Ã¶lÃ§er.

Ãzellikler ve hedef deÄiÅken arasÄ±ndaki baÄÄ±mlÄ±lÄ±ÄÄ± yakalar.

DoÄrusal olmayan iliÅkileri bile tespit edebilir.

```python
from sklearn.feature_selection import mutual_info_classif

def mi_feature_selection_with_table(X, y):
    # MI skorlarÄ±nÄ± hesapla
    mi_scores = mutual_info_classif(X, y, random_state=42)
    
    # DataFrame'e Ã§evir ve sÄ±rala
    mi_df = pd.DataFrame({
        'Feature': X.columns,
        'MI Score': mi_scores
    }).sort_values(by='MI Score', ascending=False)
    
    return mi_df
```


| MI Skoru | AnlamÄ±                      |
| -------- | --------------------------- |
| 0.0      | BaÄÄ±msÄ±z (hiÃ§ bilgi vermez) |
| 0.01â0.1 | ZayÄ±f iliÅki                |
| 0.1â0.3  | Orta dÃ¼zey iliÅki           |
| >0.3     | GÃ¼Ã§lÃ¼ iliÅki                |


```python
mi_table = mi_feature_selection_with_table(X_train_scaled, y_train)


# SonuÃ§larÄ± yazdÄ±r
print(mi_table)
```

# Model EÄitimi

Neden Birden Fazla Model KullandÄ±k?

FarklÄ± algoritmalar farklÄ± veri yapÄ±larÄ±nda daha iyi sonuÃ§ verebilir.

Modelleri karÅÄ±laÅtÄ±rarak en yÃ¼ksek baÅarÄ±yÄ± gÃ¶stereni tespit etmek istedik.

GerÃ§ek dÃ¼nyada veri Ã§eÅitliliÄi fazla olduÄu iÃ§in Ã§oklu test Ã¶nemli.

| SÄ±ra | Model AdÄ±                        | TÃ¼r / YaklaÅÄ±m                     | AÃ§Ä±klama                                                                                                                                                 |
| ---- | -------------------------------- | ---------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------- |
| 1    | **Logistic Regression**          | Lineer / Parametrik                | Basit, hÄ±zlÄ± ve yorumlamasÄ± kolay bir modeldir. DoÄrusal sÄ±nÄ±rlar Ã¼zerinden Ã§alÄ±ÅÄ±r. Ãzellikle temel sÄ±nÄ±flandÄ±rma problemleri iÃ§in uygundur.            |
| 2    | **Decision Tree**                | Karar AÄacÄ± / Kural BazlÄ±          | Veriyi if-else Åeklinde dallara ayÄ±rarak sÄ±nÄ±flandÄ±rÄ±r. Kolay yorumlanabilir. Ancak aÅÄ±rÄ± Ã¶Ärenmeye yatkÄ±ndÄ±r.                                           |
| 3    | **Random Forest**                | Ensemble / Bagging                 | Birden fazla karar aÄacÄ±nÄ±n ortalamasÄ±nÄ± alarak tahmin yapar. Genellikle yÃ¼ksek doÄruluk saÄlar. Overfitting riskini azaltÄ±r.                            |
| 4    | **Support Vector Machine (SVM)** | Margin-Maximizing / Kernel TabanlÄ± | KarmaÅÄ±k sÄ±nÄ±flandÄ±rma problemleri iÃ§in gÃ¼Ã§lÃ¼dÃ¼r. SÄ±nÄ±flar arasÄ± en iyi ayrÄ±m Ã§izgisini (hiperdÃ¼zlem) bulmaya Ã§alÄ±ÅÄ±r. KÃ¼Ã§Ã¼k veri kÃ¼melerinde etkilidir. |


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Ãzellik setlerini tanÄ±mla
X_sets = {'SelectKBest': X_kbest, 'RFE': X_rfe, 'Mutual Information': X_mi}

# Model yapÄ±landÄ±rmalarÄ±
classifiers = {
    'Logistic Regression': LogisticRegression(
        C=1.0,                        
        max_iter=1000,               
        penalty='l2',
        class_weight='balanced',     
        solver='lbfgs',              
        random_state=42
    ),

    'Decision Tree': DecisionTreeClassifier(
        max_depth=None,              
        min_samples_split=2,        
        min_samples_leaf=1,         
        class_weight='balanced',
        random_state=42
    ),

    'Random Forest': RandomForestClassifier(
        n_estimators=100,            
        max_depth=None,              
        min_samples_split=2,
        min_samples_leaf=1,
        class_weight='balanced',
        n_jobs=-1,                  
        random_state=42
    ),

    'SVM': SVC(
        C=1.0,                       
        kernel='rbf',
        gamma='scale',               
        probability=True,            
        class_weight='balanced',
        random_state=42
    )
}

# DeÄerlendirme fonksiyonu
def evaluate_models(X_sets, y, classifiers, test_size=0.2, random_state=42):
    results = []

    for set_name, X in X_sets.items():
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

        for model_name, model in classifiers.items():
            model.fit(X_train, y_train)

            # Tahminler
            y_train_pred = model.predict(X_train)
            y_test_pred = model.predict(X_test)

            # ð Metrikleri hesapla
            for split, y_true, y_pred in [('Train', y_train, y_train_pred), ('Test', y_test, y_test_pred)]:
                results.append({
                    'Feature Set': set_name,
                    'Model': model_name,
                    'Data Split': split,
                    'Accuracy': accuracy_score(y_true, y_pred),
                    'Precision': precision_score(y_true, y_pred, average='weighted', zero_division=0),
                    'Recall': recall_score(y_true, y_pred, average='weighted'),
                    'F1-Score': f1_score(y_true, y_pred, average='weighted')
                })

    return pd.DataFrame(results)

# Fonksiyonu Ã§alÄ±ÅtÄ±r
results_df = evaluate_models(X_sets, y_train, classifiers)
print("\n--- Model Performans Bilgileri (Train & Test) ---")
print(results_df)
```

```python
# En iyi modeli bul
best_accuracy = 0
best_model = None
best_feature_set = None
best_split = None

# Test seti sonuÃ§larÄ±na gÃ¶re en iyi modeli bul
test_results = results_df[results_df['Data Split'] == 'Test']

for _, row in test_results.iterrows():
    if row['Accuracy'] > best_accuracy:
        best_accuracy = row['Accuracy']
        best_model = row['Model']
        best_feature_set = row['Feature Set']
        best_split = row['Data Split']

print("\n=== En Ä°yi Model Performans Bilgileri ===")
print(f"Model: {best_model}")
print(f"Ãzellik Seti: {best_feature_set}")
print(f"Veri BÃ¶lÃ¼mÃ¼: {best_split}")
print(f"En YÃ¼ksek DoÄruluk: {best_accuracy:.4f}")

# En iyi modelin detaylÄ± metriklerini gÃ¶ster
best_model_metrics = test_results[
    (test_results['Model'] == best_model) & 
    (test_results['Feature Set'] == best_feature_set)
].iloc[0]

print("\nDetaylÄ± Metrikler:")
print(f"Precision: {best_model_metrics['Precision']:.4f}")
print(f"Recall: {best_model_metrics['Recall']:.4f}")
print(f"F1-Score: {best_model_metrics['F1-Score']:.4f}")
```

```python
# ROC Analizi iÃ§in gerekli fonksiyonlar
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
import numpy as np

def plot_roc_curves(X_sets, y, classifiers, test_size=0.2, random_state=42):
    plt.figure(figsize=(15, 10))
    
    # SÄ±nÄ±f sayÄ±sÄ±nÄ± belirle
    classes = np.unique(y)
    n_classes = len(classes)
    
    for set_name, X in X_sets.items():
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
        
        # Ãok sÄ±nÄ±flÄ± etiketleri binary formata dÃ¶nÃ¼ÅtÃ¼r
        y_test_bin = label_binarize(y_test, classes=classes)
        
        for model_name, model in classifiers.items():
            model.fit(X_train, y_train)
            
            # Her sÄ±nÄ±f iÃ§in ROC eÄrisi
            if hasattr(model, "predict_proba"):
                y_pred_proba = model.predict_proba(X_test)
                
                # Binary sÄ±nÄ±flandÄ±rma durumu iÃ§in Ã¶zel iÅlem
                if n_classes == 2:
                    # Sadece pozitif sÄ±nÄ±f olasÄ±lÄ±klarÄ±nÄ± al
                    y_pred_proba = y_pred_proba[:, 1].reshape(-1, 1)
                    y_test_bin = y_test_bin.reshape(-1, 1)
                
                # Her sÄ±nÄ±f iÃ§in ROC eÄrisi Ã§iz
                for i in range(y_test_bin.shape[1]):
                    fpr, tpr, _ = roc_curve(y_test_bin[:, i], 
                                          y_pred_proba[:, i] if n_classes > 2 else y_pred_proba[:, 0])
                    roc_auc = auc(fpr, tpr)
                    plt.plot(fpr, tpr, 
                            label=f'{model_name} ({set_name}) - Class {classes[i]} - AUC: {roc_auc:.3f}')
            else:  # SVM iÃ§in decision_function kullan
                y_pred_proba = model.decision_function(X_test)
                
                # Binary sÄ±nÄ±flandÄ±rma durumu iÃ§in Ã¶zel iÅlem
                if y_pred_proba.ndim == 1:
                    y_pred_proba = y_pred_proba.reshape(-1, 1)
                    
                # Her sÄ±nÄ±f iÃ§in ROC eÄrisi Ã§iz
                for i in range(y_test_bin.shape[1]):
                    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
                    roc_auc = auc(fpr, tpr)
                    plt.plot(fpr, tpr, 
                            label=f'{model_name} ({set_name}) - Class {classes[i]} - AUC: {roc_auc:.3f}')
    
    plt.plot([0, 1], [0, 1], 'k--')  # Referans Ã§izgisi
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC EÄrileri KarÅÄ±laÅtÄ±rmasÄ±')
    plt.legend(loc="lower right", bbox_to_anchor=(1.15, 0))
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ROC analizini gerÃ§ekleÅtir
plot_roc_curves(X_sets, y_train, classifiers)
```
